```python
import numpy as np
import pandas as pd
```

Findng Missing Values

```python
# nan --> null values --> not a number
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# find how many values are missing (nan or null) -->  df.isna() -->> gives true where value are missing otherwise false.
df.isna()
```

```python
# sum of all missing values in each columns.
df.isna().sum()
```

```python
# check that all columns have exist a null value. if exist --> Ture  , otherwise --> false
df.isna().any()
```

Removing missing Data

```python
df
```

```python
# remove all rows in which null values exist.
# dropna() --> select rows wise.
df.dropna()
```

```python
df
```

```python
# threshold --> if thresh = 3 --> it means if in each row minimum 3 non-null value exist . except these rows all rows are removed in which rows, non-null values < 3.
df.dropna(thresh=3)
```

Filling the missing data

```python
df
```

```python
# all null values replaced(filled) by 0.
# fillna() --> selsect column wise.
df.fillna(0)
```

```python
# each columns null value replaced by user defined values (different for each or same).
values = {'A':50, 'B': 100, 'C':300, 'D':400}
df.fillna(value=values)
```

```python
df
```

```python
df.mean()
```

```python
# fill missing values of each column by mean of each columns .
df.fillna(df.mean())
```
