# PANDAS SERIES
A series is a one-demensional labeled array capable of holding any data type. The axis labels are collectively called the index.

# Creating Series
there are multiple ways to create  a pandas series.

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([40,50,60])
my_dict = {'a':70, 'b':80, 'c':90}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list, index = labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(arr, index = labels)
```

```python
pd.Series(my_dict)
```
