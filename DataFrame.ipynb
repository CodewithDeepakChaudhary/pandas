# DataFrame
When we combine multiple Series then it  will became DataFrame.

```python
import numpy as np
import pandas as pd
```

1. Creating a DataFrame

```python
# using dictionary
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# using list
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]

df2 = pd.DataFrame(data_list)
```

```python
df2
```

```python
columns = ['Name', 'Age', 'City', 'Salary']
df2 = pd.DataFrame(data_list, columns = columns)
df2
```

2. Selection and indexing of columns

```python
df2
```

```python
df2['Name']   # single column select
```

```python
df2[['Name', 'City']]  # multiple column select (list of list)
```

3. Creatng a new column

```python
df2['Designation'] = ['Doctor', 'Eng.', 'Doctor', 'Eng']
```

```python
df2
```

4. Removing columns and rows

```python
# it show only copy data not deleted from original data. for deleted in original data we use (inplace = True)
# by default axis = 0 --> for horizonral(row), &&&  axis = 1 ---> for vertical(column)
df2.drop('Designation', axis = 1)    
```

```python
df2
```

```python
# deleted data from original data.  using (inplace = True)
df2.drop('Designation', axis = 1, inplace=True)
```

```python
df2
```

```python
# removing multiple columns.  (removing from original data use inplace = True)
df2.drop(['City', 'Age'], axis = 1)
```

```python
df2
```

```python
# Removing rows ---> use axis = 0  (by default axis = 0)
df2.drop(0, )
```

```python
df2.drop(0, axis = 0)
```

```python
df2
```

```python
# removing multiple rows
df2.drop([1,2,3], axis = 0)
```

```python
df2
```

7. Conditional conditions

```python
df2
```
