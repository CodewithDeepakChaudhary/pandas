```python
import numpy as np
import pandas as pd
```

```python
# DataFrame 1
employees = pd.DataFrame({
    'employee_id': [1,2,3,4,5],
    'name': ['John','Anna','Peter','Linda','Bob'],
    'department':['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2
salaries = pd.DataFrame({
    'employee_id': [1,2,3,6,7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

Merging 2 DataFrames

```python
# minimum one column should be common in both DataFrame (on = 'employee_id').
# how = 'inner' --->> bydefalut hota h.  --> means merge inner value of employee_id(rows common in both DataFrames based on common column.
pd.merge(employees,salaries, on='employee_id', how = 'inner')
```

```python
pd.merge(employees,salaries, on = 'employee_id', how = 'outer')  # outer --> merge all rows in both DataFrames based on employee_id. Where data not exist filled by NaN automatically.
```

```python
# merge common rows and left side all rows. in which left row columns data not exist filled by NaN.
pd.merge(employees,salaries, on = 'employee_id', how = 'left')
```

```python
# merge common rows and right side all rows. in which right row columns data not exist filled by NaN.
pd.merge(employees,salaries, on = 'employee_id', how = 'right')
```

Concatination of 2 DataFrames

```python
df1 = pd.DataFrame({
    'A': ['A0','A1','A2'],
    'B': ['B0','B1','B2'],
    'C': ['C0','C1','C2']
})

df2 = pd.DataFrame({
    'A': ['A3','A4','A5'],
    'B': ['B3','B4','B5'],
    'C': ['C3','C4','C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])  # bydefault column based cancat (axis = 0)
```

```python
pd.concat([df1,df2], axis = 1)  # row based concat
```

Joining 2 DataFrames

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index = [1,2,3])

df2 = pd.DataFrame({
    'score': [85,90, 75]
}, index = [2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df1.join(df2, how = 'outer')
```

```python
df1.join(df2, how='right')
```
